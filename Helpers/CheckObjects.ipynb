# Verificando Arquivos

```python
minio = MinioClient()
```

```python
buckets = minio.list_buckets();buckets
```

```python
files = minio.list_objects(buckets[0]);files
```

```python
objects = {
    PurePosixPath(file).stem:minio.get_object(
        bucket_name=buckets[0],
        object_name=file
        ) 
        for file in minio.list_objects(buckets[0])[1:]
    }
```

```python
# Lendo um parquet:
df = pl.read_parquet(objects["terceirizados_2022_setembro"]);df.head()
```

```python
# lendo os parquets:
parquet_dfs = []

for name, buffer in objects.items():
    if name.endswith("bronze"):
        continue

    buffer.seek(0)
    parquet_dfs.append(pl.read_parquet(buffer))
```

```python
import duckdb
import tempfile
import polars as pl
```

```python
duck_dfs = []

duck_buffer = minio.get_object(
    bucket_name=buckets[0],
    object_name=files[0]
)

duck_buffer.seek(0)

# cria arquivo temporário SEM manter aberto
tmp = tempfile.NamedTemporaryFile(suffix=".duckdb", delete=False)
tmp.write(duck_buffer.read())
tmp.close()   # 🔥 ESSENCIAL NO WINDOWS

# agora pode abrir com duckdb
con = duckdb.connect(tmp.name)

table_names = [
    row[0]
    for row in con.execute("SHOW TABLES").fetchall()
]

for table in table_names:
    df = con.execute(f"SELECT * FROM {table}").pl()
    duck_dfs.append(df)

con.close()
```

```python
parquet_dfs[1].columns
```

```python
results = []

for p_df, d_df in zip(parquet_dfs, duck_dfs):
    equal = (
        p_df.join(d_df, on="id", how="anti").height == 0
        and
        d_df.join(p_df, on="id", how="anti").height == 0
    )
    results.append(equal)

print(results)
```
